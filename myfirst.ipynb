```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(0)

# Create a DataFrame with random data
num_rows = 10
random_data = np.random.rand(num_rows)
df = pd.DataFrame(random_data, columns=['Values'])

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['Values'], color='blue')
plt.title('Bar Chart of Random Values')
plt.xlabel('Index')
plt.ylabel('Values')
plt.xticks(df.index)  # Set x-ticks to be the index of the DataFrame
plt.show()
```
